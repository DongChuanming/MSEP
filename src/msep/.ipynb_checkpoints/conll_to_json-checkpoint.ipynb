```python
def sample_selection(s, n):
    ns=list()
    ls=sorted(s, key=lambda a:len(a["text"]), reverse=True)
    total=len(ls)
    if n<total:
        times=round(total/n)
        for i in range(n):
            try:
                ns.append(ls[i*times])
            except:
                return ns
    else:
        return s
    return ns
```

```python
import json
#lk=['text', 'meta', 'tokens_parser', 'spans', 'tokens', '_input_hash', '_task_hash', '_view_id', 'answer', '_timestamp']
with open("preannotation_complet.txt", "r", encoding="utf-8") as pre:
    output=open("prodigy_preannotation.jsonl", "w", encoding="utf-8")
    ID_ENTREPOT=""
    ID_PAT=""
    CODE=""
    line_count=0
    #dic={'text':"", 'meta':"", 'tokens_parser':"", 'spans':"", 'tokens':"", '_input_hash':"", '_task_hash':"", '_view_id':"", 'answer':"", '_timestamp':""}
    
    lp=pre.readlines()[1:]
    tk_b=0
    tk_e=0
    for l in lp:
        l=l.strip()
        if l:
            if "#ID_ENTREPOT : " in l:
                ID_ENTREPOT=float(l.replace("#ID_ENTREPOT : ", "").strip())
            elif "#ID_PAT : " in l:
                ID_PAT=float(l.replace("#ID_PAT : ", "").strip())
            if ID_ENTREPOT:
                CODE=d_idcode[ID_ENTREPOT]
            
            if l[0]!="#":
                ll=l.split("\t")
                line_count+=1
                #if len(ll) != 8:
                #    print(ll)
                txt=ll[0]
                tk_e=tk_b+len(txt)
                dic_preann=dict()
                dic_preann["taba"]=ll[1]
                dic_preann["prof"]=ll[2]
                dic_preann["diab"]=ll[3]
                dic_preann["hyper"]=ll[4]
                dic_preann["cardia"]=ll[5]
                dic_preann["copd"]=ll[6]
                dic_preann["fam"]=ll[7]
                key_count=0
                for k,v in dic_preann.items():
                    key_count+=1
                    dic={"text":"", "meta":"", "tokens_parser":"", "spans":"", "tokens":"", "_input_hash":"", "_task_hash":"", "_view_id":"", "answer":"", "_timestamp":""}
                    dic["text"]=txt
                    dic["meta"]={"ID_PAT": ID_PAT, "ID_ENTREPOT": ID_ENTREPOT, "CODE": CODE, "annotator": 5, "annotator2": null}
                    dic["tokens_parser"]=[{"start_char": tk_b, "end_char": tk_e, "text": txt, "start": 0, "end": len(txt), "id": 0}]
                    dic["spans"]=[{"start": tk_b, "end": tk_e, "label": v, "token_start": 0, "token_end": 0}]
                    dic["tokens"]= [{"text": txt, "start": 0, "end": len(txt), "id": 0, "ws": false}]
                    dic["_input_hash"]=-line_count
                    dic["_task_hash"]=-key_count
                    dic["_view_id"]="ner_manual"
                    dic["answer"]="accept"
                    dic["_timestamp"]=line_count*10+key_count
                    output.write(json.dumps(dic, ensure_ascii=False).strip()+"\n")
                    
                
                    
                    
                
                
                tk_b=tk_e+1
                
                
            
            
                
                
            
```

```python
ldp=list()
import json
with open("projet_poline/poline_base_1.jsonl", "r", encoding="utf-8") as f:
    lf=f.readlines()
    for l in lf:
        #try:
        d=json.loads(l)
        ldp.append(d)
```

```python
with open("projet_poline/poline_base_1.csv", "w", encoding="utf-8") as w:
    w.write("anteur_année"+"\t"+"titre"+"\t"+"equation"+"\t"+"text"+"\t"+'CRIT_1_INDISPONIBLE'+"\t"+'CRIT_2_HS'+"\t"+'CRIT_PAS_CONFORME'+"\t"+'CRIT_4'+"\t"+"commentaire"+"\n")
    
    for d in ldp:
        line=d["meta"]['auteur_année'] +"\t"+ d["meta"]['titre'] +"\t"+d["meta"]['equation'] +"\t"+d["text"]+"\t"  
        lbs=d["accept"]
        annot=["","","",""]
        if lbs:
            for i in lbs:
                annot[int(i)]="1"
        commentaire = ""        
        if "user_input" in d:
            commentaire=d["user_input"]
        line+="\t".join(annot)+"\t"+commentaire+"\n"
        w.write(line)
            
```
